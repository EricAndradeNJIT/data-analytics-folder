```python
#import packages

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
#read and rename

cancer = pd.read_csv('cancer.csv')

cancer.head()
```

```python
#convert diagnosis into numeric data type

cancer.loc[cancer['diagnosis'] == 'M', 'cancer_present'] = 1
cancer.loc[cancer['diagnosis'] == 'B', 'cancer_present'] = 0
```

```python
#split data into target variable and feature of interest

y = cancer[['cancer_present']]

X = cancer[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 
            'concavity_mean', 'concave points_mean',	'symmetry_mean',	'fractal_dimension_mean']]
```

```python
X.describe()
```

```python
#Split data into training/validation dataset and test dataset

X_train_val, X_test, y_train_val, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
#Split training/validation datasets into traning and validation sets

X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.3333, random_state=42) 
```

```python
#look at measures in trainig data

X_train.describe()
```

```python
#check for missing data

X.isna().sum()
```

```python
#impute data

imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')

X_train_ = imp_mean.fit_transform(X_train)
                                  
X_train = pd.DataFrame(X_train_)
```

```python
#check for missing data

X_train.isna().sum()
```

```python
#Standardize data

std = StandardScaler()

X_train_std_ = std.fit_transform(X_train)

X_train_std = pd.DataFrame(X_train_std_)

print(round(X_train_std.describe(),2))
```

```python
# fit scaler on training data
norm = MinMaxScaler().fit(X_train)
```

```python
# transform training data
X_train_norm_ = norm.transform(X_train)

X_train_norm = pd.DataFrame(X_train_norm_)

print(round(X_train_norm.describe(),2))
```
