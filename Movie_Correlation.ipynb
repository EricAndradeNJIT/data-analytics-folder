```python
#Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#Read data

df = pd.read_csv('movies.csv') 
```

```python
df.head()
```

```python
#Find missing data

for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col, pct_missing))
```

```python
#Delete missing values
df = df.dropna()
```

```python
#Check if missing values were deleted
for col in df.columns:
  pct_missing = np.mean(df[col].isnull())
  print('{} - {}%'.format(col, pct_missing))
```

```python
#check for data types
df.dtypes
```

```python
#change column data types

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')

df['votes'] = df['votes'].astype('int64')
```

```python
df
```

```python
#correct year column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
#order data by gross column

df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#check and drop duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#looking for correlation using scatterplots

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross')

plt.xlabel('Budget')
plt.ylabel('Gross')

plt.show()
```

```python
df
```

```python
#plot budget vs gross using sns

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})
```

```python
#checking correlations

df.corr()
```

```python
df.corr(method='pearson')
```

```python
df.corr(method= 'kendall')
```

```python
df.corr(method= 'spearman')
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
#Look for correaltion in string column
#converting string into numeric data

df_numbered = df

for col_name in df_numbered.columns:
  if(df_numbered[col_name].dtype == 'object'):
    df_numbered[col_name] = df_numbered[col_name].astype('category')
    df_numbered[col_name] = df_numbered[col_name].cat.codes

df_numbered
```

```python
correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numbered.corr()
```

```python
correlation_mat = df_numbered.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs.head(5)
```

```python
sorted_pairs = corr_pairs.sort_values(ascending=False)

sorted_pairs.head()
```

```python
high_correlation = sorted_pairs[(sorted_pairs) > 0.5]

high_correlation
```
